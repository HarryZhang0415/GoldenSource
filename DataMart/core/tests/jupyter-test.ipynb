```python
from datamart_core.app.model.obbject import OBBject
import requests

url = "http://localhost:8001/api/v1/equity/price/historical?chart=false&provider=yfinance&symbol=AAPL&interval=1d&start_date=2024-01-01&end_date=2024-01-31&adjustment=splits_only&extended_hours=false&adjusted=false&use_cache=true&timezone=America%2FNew_York&source=realtime&sort=asc&limit=49999&include_actions=true&prepost=false"

res = requests.get(url)
```

```python
AAPL_1 = OBBject.from_response(res)
```

```python
AAPL_1.to_dataframe()
```
